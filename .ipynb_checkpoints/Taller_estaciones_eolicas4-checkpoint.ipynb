```python
import time,datetime
import pandas
def archivo (file):
    estacion4 = open('C:/Users/VERONICA/Desktop/SENA/5t/PYTHON/files/06/Estacion4.csv', 'r').readlines()
    estacion4 = estacion4[1:]
    estacion4 = [row[0:-1] for row in estacion4]
    estacion4 = [row.split(';') for row in estacion4]
    return estacion4
```

```python
tabla4= archivo('C:/Users/VERONICA/Desktop/SENA/5t/PYTHON/files/06/Estacion4.csv')
```

```python
[[row[0].lower()]for row in tabla4]
tabla4=[[str('Estacion4')]+ row for row in tabla4]
```

```python
def cambio(s):
    return datetime.datetime.strptime(s,'%d/%m/%y').strftime('%Y/%m/%d')
```

```python
tabla4= [[row[0],cambio(row[1]),row[2],row[3],row[4]] for row in tabla4]
```

```python
def YEAR(s):
    return datetime.datetime.strptime(s,'%Y/%m/%d').strftime('%Y')
```

```python
tabla4=[[YEAR(row[1])]+ row for row in tabla4]
```
