```python
import time,datetime
import pandas
def archivo (file):
    estacion2 = open('C:/Users/VERONICA/Desktop/SENA/5t/PYTHON/files/06/Estacion2.csv', 'r').readlines()
    estacion2 = estacion2[1:]
    estacion2 = [row[0:-1] for row in estacion2]
    estacion2 = [row.split(';') for row in estacion2]
    return estacion2
```

```python
tabla2= archivo('C:/Users/VERONICA/Desktop/SENA/5t/PYTHON/files/06/Estacion2.csv')
```

```python
[[row[0].lower()]for row in tabla2]
tabla2=[[str('Estacion2')]+ row for row in tabla2]
```

```python
def cambio2(s):
    return datetime.datetime.strptime(s,'%d/%m/%y').strftime('%Y/%m/%d')
```

```python
tabla2= [[row[0],cambio2(row[1]),row[2],row[3],row[4]] for row in tabla2]
```

```python
def YEAR(s):
    return datetime.datetime.strptime(s,'%Y/%m/%d').strftime('%Y')
```

```python
tabla2=[[YEAR(row[1])]+ row for row in tabla2]
```

```python
tabla2=[[row[1],row[2],row[0],row[3],row[4],row[5]]for row in tabla2]
```

```python
def MONTH(s):
    return datetime.datetime.strptime(s,'%Y/%m/%d').month
```

```python
tabla2=[[MONTH(row[1])]+ row for row in tabla2]
```

```python
tabla2=[[row[1],row[2],row[3],row[0],row[4],row[5],row[6]]for row in tabla2]
```

```python
def DAY(s):
    return datetime.datetime.strptime(s,'%Y/%m/%d').strftime('%d')
```

```python
tabla2=[[DAY(row[1])]+ row for row in tabla2]
```

```python
tabla2=[[row[1],row[2],row[3],row[4],row[0],row[5],row[6],row[7]]for row in tabla2]
```

```python
tabla2=[[HORA(row[5])]+ row for row in tabla2]
```
